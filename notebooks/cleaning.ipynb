```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
contact_csv=pd.read_csv('../data/contact_df.csv',index_col=0)
engine = create_engine('sqlite:///../data/lobbyist.db')

contact_csv.head(20)
```

```python
Officials=contact_csv['City Official'].unique().tolist()
Officials
```

```python

def split_data(entry):
    if pd.isna(entry):
        return None, None
    elif ',' in entry:
        name, title = entry.split(',', 1)
        return title.strip(), name.strip()
    else:
        return entry.strip(), None

contact_csv['office_name'], contact_csv['official_name'] = zip(*contact_csv['City Official'].apply(split_data))
contact_csv = contact_csv.drop(columns=['City Official'])
```
